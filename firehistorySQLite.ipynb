```python
import pandas as pd
import requests
import sqlite3
import json

conn = sqlite3.connect('project.db')
c = conn.cursor()


c.execute("DROP TABLE IF EXISTS firedata;")
sqltablefire = """
        CREATE TABLE firedata ( 
        ID BLOB PRIMARY KEY, 
        acq_date NUMERIC, 
        bright_t31 NUMERIC, 
        brightness NUMERIC, 
        daynight VARCHAR(100),
        frp NUMERIC,
        instrument VARCHAR(100),
        latitude NUMERIC,
        longitude NUMERIC,
        satellite VARCHAR(100));"""

c.execute(sqltablefire)

```

```python
c.execute("DROP TABLE IF EXISTS firehistorical;")
sqltablefirehist = """
        CREATE TABLE firehistorical ( 
        ID BLOB PRIMARY KEY, 
        Date NUMERIC,
        Name VARCHAR(100),
        States_territories VARCHAR(100),
        AreaBurned_ha NUMERIC, 
        AreaBurned_acres NUMERIC,
        Fatalities NUMERIC,
        PropertiesDamaged_HomesDestroyed NUMERIC,
        Year NUMERIC);"""
c.execute(sqltablefirehist)
c.execute("DROP TABLE IF EXISTS rainfall;")
sqltablerainfall = """
        CREATE TABLE rainfall ( 
        ID BLOB PRIMARY KEY, 
        Year NUMERIC,
        Avg_Annual_Temp NUMERIC,
        Avg_Annual_Rainfall NUMERIC,
        temp_differnce NUMERIC,
        rainfall_difference NUMERIC
        );"""


c.execute(sqltablerainfall)
          
          
          
```

```python
import json
import urllib.request


url = "http://australia-fire-api-dashboard.herokuapp.com/api/v1.0"
modis = '/fires_modis'
historical = '/fires_historical'
rainfall = '/aus_temp_rainfall'

response = requests.get(url+modis)
modisdata = response.json()
#print(modisdata)

df = pd.read_json(url+modis, orient='index')
df[0]
result = c.fetchall()
print(result)






data = urllib.request.urlopen(url+modis).read().decode()

obj = json.loads(data)

for child in obj['result']:
    #print(child)
    c.execute("Insert into firedata values (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", 
                       (child['id'],child['acq_date'], child['bright_t31'], child['brightness'], child['daynight'], 
                       child['frp'], child['instrument'],child['latitude'],child['longitude'],
                        child['satellite']))
    conn.commit()
        

```

```python
data = urllib.request.urlopen(url+historical).read().decode()

obj = json.loads(data)
#print(obj['result'])


for child in obj['result']:
    #print(child)
    c.execute("Insert into firehistorical values (?, ?, ?, ?, ?, ?, ?, ?, ?)", 
                       (child['id'],child['date'], child['name'], child['state'], child['area_burned_ha'], 
                       child['area_burned_acres'], child['fatalities'],child['homes_destroyed'],child['year']))
    conn.commit()

  
```

```python
data = urllib.request.urlopen(url+rainfall).read().decode()

obj = json.loads(data)
#print(obj['result'])


for child in obj['result']:
    #print(child)
    c.execute("Insert into rainfall values (?, ?, ?, ?, ?, ?)", 
                       (child['id'], child['year'], child['avg_annual_temp'], child['avg_annual_rainfall'], 
                       child['temp_difference'], child['rainfall_difference']))
    conn.commit()
```
