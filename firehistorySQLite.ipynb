```python
import pandas as pd
import requests
import sqlite3
import json

conn = sqlite3.connect('project.db')
c = conn.cursor()


c.execute("DROP TABLE IF EXISTS firedata;")
sqltablefire = """
        CREATE TABLE firedata ( 
        ID BLOB PRIMARY KEY, 
        acq_date NUMERIC, 
        bright_t31 NUMERIC, 
        brightness NUMERIC, 
        daynight VARCHAR(100),
        frp NUMERIC,
        instrument VARCHAR(100),
        latitude NUMERIC,
        longitude NUMERIC,
        satellite VARCHAR(100));"""

c.execute(sqltablefire)

```

```python
c.execute("DROP TABLE IF EXISTS firehistorical;")
sqltablefirehist = """
        CREATE TABLE firehistorical ( 
        ID BLOB PRIMARY KEY, 
        Date NUMERIC,
        Name VARCHAR(100),
        States_territories VARCHAR(100),
        AreaBurned_ha NUMERIC, 
        AreaBurned_acres NUMERIC,
        Fatalities NUMERIC,
        PropertiesDamaged_HomesDestroyed NUMERIC,
        Year NUMERIC);"""
c.execute(sqltablefirehist)
c.execute("DROP TABLE IF EXISTS rainfall;")
sqltablerainfall = """
        CREATE TABLE rainfall ( 
        ID BLOB PRIMARY KEY, 
        Year NUMERIC,
        Avg_Annual_Temp NUMERIC,
        Avg_Annual_Rainfall NUMERIC,
        temp_differnce NUMERIC,
        rainfall_difference NUMERIC
        );"""


c.execute(sqltablerainfall)
          
          
          
```

```python
#obj = json.loads('Bush_Fire_Prone_Areas_2021_OBRM_019_WA_GDA2020_Public.geojson')
from flask import jsonify
import geojson
with open('Bush_Fire_Prone_Areas_2021_OBRM_019_WA_GDA2020_Public.geojson') as f:
    gj = geojson.load(f)

for child in gj['features']:
    coor = json.dumps(child['geometry']['coordinates'][0])
    c.execute("INSERT INTO Fireprone(Coordinates) VALUES(?)",[coor])
    ID = json.dumps(child['properties']['object_id'])
    c.execute("INSERT INTO  Fireprone(ID) VALUES(?)", [ID])
    lga = json.dumps(child['properties']['lga'])
    c.execute("INSERT INTO Fireprone(lga) VALUES(?)",[lga])
    designation = json.dumps(child['properties']['designation'])
    c.execute("INSERT INTO Fireprone(Designation) VALUES(?)",[designation])
    typ = json.dumps(child['properties']['type'])
    c.execute("INSERT INTO Fireprone(Type) VALUES(?)", [typ])
    designationdate = json.dumps(child['properties']['designationdate'])
    c.execute("INSERT INTO Fireprone(designationdate) VALUES(?)", [designationdate])
    st_area_shape_ = json.dumps(child['properties']['st_area_shape_'])
    c.execute("INSERT INTO Fireprone(Area) VALUES(?)", [st_area_shape_])
    st_perimeter_shape_ = json.dumps(child['properties']['st_perimeter_shape_'])
    c.execute("INSERT INTO Fireprone(Perimeter) VALUES(?)", [st_perimeter_shape_])
                    


    conn.commit()

statement = '''SELECT * FROM Fireprone'''
  
c.execute(statement)
  
print("All the data")
output = c.fetchall()
count = 0
while count <5:
    for row in output:
        count = count+1
        print(row)
```
